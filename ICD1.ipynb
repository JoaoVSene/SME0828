####Exercício 2:

```python
# Leitura do DataFrame
data = pd.read_csv('data/iris-with-errors.csv', header=(0))
print("Número de linhas e colunas:",data.shape)

data = data.replace('?', np.nan) # Substitui '?' por NaN

# Convertemos para o formato Numpy e ignoramos a última coluna (classe)
X = np.array(data[data.columns[0:data.shape[1]-1]], dtype = float)

# Calculamos a mediana de cada atributo e armazenamos em um vetor:

averages = np.nanmedian(X, axis = 0) # Usamos a função nanmedian que calcula a mediana ignorando os Nan

for i in np.arange(0, X.shape[0]):
    for j in np.arange(0, X.shape[1]):
        if(np.isnan(X[i,j]) == True): # Verificar se é uma entrada com Nan
            X[i,j] = averages[j] # Insere a mediana
print(X)

data_new = pd.DataFrame(X, columns = data.columns[0:data.shape[1]-1]) # Convertendo para DataFrame

data_new.head(25)
```

####Exercício 3

```python
# Biblioteca para pré-processamento de dados:
from sklearn.preprocessing import MinMaxScaler 
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('data/iris.csv', header=(0)) # Leitura do DataFrame

# Normalização:

scaler = MinMaxScaler(feature_range=(0, 1)) # Prepara a função para transformar os dados
data.iloc[:,:-1] = scaler.fit_transform(data.iloc[:,:-1]) # Realiza a normalização e coloca em um novo vetor

X = np.array(data[data.columns[0:data.shape[1]-1]])

print('Mínimo dos atributos:', np.amin(X, axis=0))
print('Máximo dos atributos:', np.amax(X, axis=0))

print("Distribuição:")

data.hist(column='sepal_length')
data.hist(column='sepal_width')
data.hist(column='petal_length')
data.hist(column='petal_width')

plt.show()
```

```python
# Padronização:

scaler = StandardScaler().fit(data.iloc[:,:-1])
# Realiza a padronização (média=0, variância = 1)
data.iloc[:,:-1] = scaler.transform(data.iloc[:,:-1])

X = np.array(data[data.columns[0:data.shape[1]-1]])

for i in np.arange(0,X.shape[1]):
    print('Media da coluna', i, 'é igual a: ', np.mean(X[:,i]))
    print('Desvio padrão da coluna', i, 'é igual a: ', np.std(X[:,i]),'\n')

print("Distribuição:")

data.hist(column='sepal_length')
data.hist(column='sepal_width')
data.hist(column='petal_length')
data.hist(column='petal_width')

plt.show()
```

```python
data = pd.read_csv('data/iris.csv', header=(0))

classes = data[data.columns[-1]]
print(classes)

N = 10
# classes
cl = np.unique(classes)
X = np.array(data)
Xnew = []
cls = np.array(data[data.columns[-1]])
for i in np.arange(0, len(cl)):
    a = np.argwhere(cls == cl[i])
    # seleciona os índices nos dados de forma aleatória
    inds = np.random.choice(a[:,0],N,replace=False)
    Xnew.append(X[inds,:])
Xnew = np.array(Xnew)

print('Dados obtidos a partir da amostragem')
print(Xnew)
```

```python
data = pd.read_csv('data/iris.csv', header=(0))
```

```python
sns.pairplot(data)
plt.show()
```

```python
# Usando distância interquantil para identificar os outliers
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

v = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1) # Guarda os outliers
data = data.drop(data.index[list(v)], axis=0) # Retira os outliers
data.head(25)
```
